```python
import pandas as pd

dados = pd.read_csv('aluguel.csv', sep=';')
dados
```

```python
# 1 a 2
# 3 a 4
# 5 a 6
# 7 ou mais
# quartos

classes = [0, 2, 4, 6, 100]
quartos = pd.cut(dados['Quartos'], classes)
quartos
```

```python
quartos.value_counts()
```

```python
label = ['1 a 2 quartos', '3 a 4 quartos', '5 a 6 quartos', '7 quartos ou mais']
quartos = pd.cut(dados['Quartos'], classes, labels=label, include_lowest=True)
quartos.value_counts()
```
