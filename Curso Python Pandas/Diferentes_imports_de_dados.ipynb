# Web Scrapping com BeautifulSoup

```python
# Importa as requests do Python e o BeautifulSoup
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Recebe o html pela request get
url='https://www.federalreserve.gov/releases/h3/current/default.htm'
response = requests.get(url)
html = response.content

# Entrega o html para o BeautifulSoup e pede pelas tabelas do site
soup = BeautifulSoup(html, 'html.parser')
table = soup.findAll('table')

# Incorpora as tabelas em um novo html
html_file = f'<html><body>{table}</body></html>'

# Faz a leitura do novo HTML com o pandas
dados_html_2 = pd.read_html(html_file)
dados_html_2[0]
```
