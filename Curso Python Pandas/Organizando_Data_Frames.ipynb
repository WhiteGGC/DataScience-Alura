```python
import pandas as pd
```

```python
data = [(1, 2, 3), (4, 5, 6), (7, 8, 9)]
data
```

```python
index = list('321')
index
```

```python
df = pd.DataFrame(data, index, list('ZYX'))
df
```

```python
df.sort_index(inplace = True)
df
```

```python
df.sort_index(inplace = True, axis = 1)
df
```

```python
df.sort_values(inplace = True, by = 'X')
df
```

```python
df.sort_values(inplace = True, by = '3', axis = 1)
df
```
